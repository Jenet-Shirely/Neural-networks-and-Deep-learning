```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.array([
    [100, 100, 100, 100, 100, 100],
    [100, 150, 150, 150, 150, 100],
    [100, 150, 200, 200, 150, 100],
    [100, 150, 200, 200, 150, 100],
    [100, 150, 150, 150, 150, 100],
    [100, 100, 100, 100, 100, 100]
], dtype=np.float32)

plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.colorbar()
plt.show()
```

```python
kernel = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
], dtype=np.float32)
```

```python
def apply_convolution(image, kernel, stride=1, padding=0):
    if padding > 0:
        image = np.pad(image, pad_width=padding, mode='constant', constant_values=0)
    i_height, i_width = image.shape
    k_height, k_width = kernel.shape


    o_height = (i_height - k_height) // stride + 1
    o_width = (i_width - k_width) // stride + 1

    output = np.zeros((o_height, o_width), dtype=np.float32)


    for i in range(0, o_height):
        for j in range(0, o_width):
            region = image[i:i+k_height, j:j+k_width]
            output[i, j] = np.sum(region * kernel)

    return output
```

```python
filtered_image = apply_convolution(image, kernel, padding=1)
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(filtered_image, cmap='gray')
plt.title("Filtered Image (Vertical Edges)")

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = "/content/WhatsApp Image 2025-03-31 at 16.41.57_f79f4039.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)


sobel_vertical = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
], dtype=np.float32)


filtered_image = cv2.filter2D(image, -1, sobel_vertical)


plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(filtered_image, cmap='gray')
plt.title("Vertical Edge Detection")

plt.tight_layout()
plt.show()
```
